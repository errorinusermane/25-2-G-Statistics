```python
import torch
import torch.nn as nn

X1 = torch.tensor([0.83, -0.18, 0.27, 0.32, 0.25, -0.73, 0.19, -0.08, -0.47, -0.28,
                   -0.23, 0.25, 0.15, 0.92, 1.27, -0.50, 0.05, -0.74, -0.06, 0.70])
X2 = torch.tensor([0.63, -1.43, -1.35, 0.70, -1.48, 0.06, 1.14, 0.96, -1.65, 0.44,
                   0.62, 0.47, -0.08, 1.03, 1.24, -0.54, 0.08, 0.85, -1.53, 0.66])
Y  = torch.tensor([0.50, 1.34, 1.39, 0.40, 1.20, 0.11, -0.21, -0.26, 1.18, 0.59,
                   0.13, 0.17, 0.55, 0.32, 0.19, 0.46, 0.32, -0.35, 1.35, 0.20])

# (20,2) 형태의 X 텐서로 결합
X = torch.stack((X1, X2), dim=1)
n = len(Y)
```

```python
# ================================================================
# (i) Optimizer 사용 (직접 모델 파라미터 정의)
# ================================================================

# 초기화
beta0 = torch.randn(1, requires_grad=True)
beta1 = torch.randn(1, requires_grad=True)
beta2 = torch.randn(1, requires_grad=True)

optimizer = torch.optim.SGD([beta0, beta1, beta2], lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()
    Y_hat = beta0 + beta1*X1 + beta2*X2
    loss = torch.mean((Y - Y_hat)**2)
    loss.backward()
    optimizer.step()

print("----- (i) Optimizer with manual parameters -----")
print(f"β0 = {beta0.item():.4f}, β1 = {beta1.item():.4f}, β2 = {beta2.item():.4f}")
print(f"MSE = {loss.item():.6f}\n")
```

```python
# ================================================================
# (ii) torch.nn.Linear(bias=True) + optimizer 사용
# ================================================================

model = nn.Linear(2, 1, bias=True)   # 입력 2, 출력 1
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()
    Y_hat = model(X).squeeze()        # (20,)으로 맞추기
    loss = torch.mean((Y - Y_hat)**2)
    loss.backward()
    optimizer.step()

print("----- (ii) nn.Linear(bias=True) + optimizer -----")
print(f"Weight = {model.weight.detach().numpy()}")
print(f"Bias   = {model.bias.detach().numpy()}")
print(f"MSE = {loss.item():.6f}\n")
```

```python
# ================================================================
# (iii) torch.nn.Linear(bias=True) + optimizer 없이 직접 gradient 업데이트
# ================================================================

model = nn.Linear(2, 1, bias=True)
lr = 0.01

for epoch in range(1000):
    Y_hat = model(X).squeeze()
    loss = torch.mean((Y - Y_hat)**2)
    loss.backward()

    with torch.no_grad():
        for param in model.parameters():
            param -= lr * param.grad
    model.zero_grad()

print("----- (iii) nn.Linear(bias=True) + manual gradient update -----")
print(f"Weight = {model.weight.detach().numpy()}")
print(f"Bias   = {model.bias.detach().numpy()}")
print(f"MSE = {loss.item():.6f}\n")
```

```python
# ================================================================
# (iv) torch.nn.Linear(bias=False) + optimizer 사용
# ================================================================

model = nn.Linear(2, 1, bias=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

for epoch in range(1000):
    optimizer.zero_grad()
    Y_hat = model(X).squeeze()
    loss = torch.mean((Y - Y_hat)**2)
    loss.backward()
    optimizer.step()

print("----- (iv) nn.Linear(bias=False) + optimizer -----")
print(f"Weight = {model.weight.detach().numpy()}")
print(f"MSE = {loss.item():.6f}")
```
